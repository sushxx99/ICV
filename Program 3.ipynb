```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display an image using matplotlib
def display_image(image, title="Image"):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Read the input image
image = cv2.imread(r"C:\Users\Student\Downloads\peacock.jpg")  # Replace 'input_image.jpg' with your image file path

# Check if image has been successfully loaded
if image is None:
    print("Error: Could not open or find the image.")
else:
    # Display the original image
    print("Original Image:")
    display_image(image)

    # Applying erosion
    kernel_erode = np.ones((3, 3), np.uint8)
    image_erode = cv2.erode(image, kernel_erode)
    print("Erosion:")
    display_image(image_erode)

    # Applying dilation
    kernel_dilate = np.ones((2, 2), np.uint8)
    image_dilate = cv2.dilate(image, kernel_dilate, iterations=1)
    print("Dilation:")
    display_image(image_dilate)

    # Applying opening
    kernel_open = np.ones((5, 5), dtype=np.uint8)
    image_opening = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel_open)
    print("Opening:")
    display_image(image_opening)

    # Applying closing
    kernel_close = np.ones((5, 5), dtype=np.uint8)
    image_closing = cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel_close)
    print("Closing:")
    display_image(image_closing)
```
