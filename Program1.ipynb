```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Define the path to your image
image_path = r"C:\Users\Student\Downloads\lion.jpg"  # Adjust the path to your image

# Read the image
img = cv2.imread(image_path)

# Function to display an image using matplotlib
def plot_image(img, title=None):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    if title:
        plt.title(title)
    plt.show()

# Display the original image
plot_image(img, title='Original Image')

# Applying different filters

# Kernel Blurring using filter2D()
kernel_25 = np.ones((25, 25), np.float32) / 625.0
output_kernel = cv2.filter2D(img, -1, kernel_25)
plot_image(output_kernel, title='Kernel Blurring (filter2D)')

# Box Filter (Simple Blur)
output_box = cv2.boxFilter(img, -1, (25, 25))
plot_image(output_box, title='Box Filter (Simple Blur)')

# Gaussian Blur
output_gaus = cv2.GaussianBlur(img, (25, 25), 0)
plot_image(output_gaus, title='Gaussian Blur')

# Median Blur (Reduction of noise)
output_med = cv2.medianBlur(img, 25)
plot_image(output_med, title='Median Blur')

# Bilateral Filtering (Reduction of noise + Preserving edges)
output_bil = cv2.bilateralFilter(img, 9, 75, 75)
plot_image(output_bil, title='Bilateral Filtering')

# Canny Edge Detection
edges = cv2.Canny(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY), 100, 200)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```
