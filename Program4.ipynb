```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display an image using matplotlib
def display_image(image, title="Image"):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.axis('off')
    plt.show()

# Read the input image
image_rgb = cv2.imread(r"C:\Users\Student\Downloads\panda.jpg")  # Replace 'input_image.jpg' with your image file path

# Check if image has been successfully loaded
if image_rgb is None:
    print("Error: Could not open or find the image.")
else:
    # Display the original RGB image
    print("Original Image in RGB color space:")
    display_image(image_rgb)

    # Convert RGB to HSV
    image_hsv = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2HSV)
    print("Image in HSV color space:")
    display_image(image_hsv)

    # Convert RGB to CMYK
    image_normalized = image_rgb / 255.0
    C = 1 - image_normalized[:, :, 2]  # Cyan
    M = 1 - image_normalized[:, :, 1]  # Magenta
    Y = 1 - image_normalized[:, :, 0]  # Yellow
    K = np.min([C, M, Y], axis=0)  # Key (Black)
    image_cmyk = np.dstack((C, M, Y, K))
    image_cmyk = (image_cmyk * 255).astype(np.uint8)  # Scale to 0-255 range
    print("Image in CMYK color space:")
    display_image(image_cmyk)

    # Convert RGB to Grayscale
    image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
    print("Image in Grayscale color space:")
    display_image(image_gray)
```
